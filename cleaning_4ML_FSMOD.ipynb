```python
# DF for target 'Severity' and features 'Start Time'
df_time2=pd.DataFrame(
    data=df,columns=['Severity','Start_Time'],index=df.index.copy())
df_time2
```

```python
df_time2.dtypes
```

```python
df_time2['Start_Time'] = pd.to_datetime(df_time2['Start_Time'])
df_time2
```

```python
df_time2['Hour'] = df_time2['Start_Time'].dt.hour
df_time2
```

```python
df_time2.dtypes
```

```python
df_time2['Month'] = df_time2['Start_Time'].dt.month
df_time2
```

```python
df_time2.dtypes
```

```python
#df_time2.to_csv(r'C:\BootcampUCB\Final Project\df_hour_month.csv', index = False)
```

```python
df_time2['Grouped_Hour'] = df_time2['Hour']
df_time2
```

```python
df_time2['Grouped_Month'] = df_time2['Month']
df_time2
```

```python
df_time2['Grouped_Severity'] = df_time2['Severity']
df_time2
```

```python
# Replacing Severity level 1 and 2 to 0
df_time2['Grouped_Severity']=df_time2['Grouped_Severity'].replace([1,2],0)
df_time2
```

```python
# Replacing Severity level 3 and 4 to 1
df_time2['Grouped_Severity']=df_time2['Grouped_Severity'].replace([3,4],1)
df_time2
```

```python
#Binning Grouped_Hour into 6 hour increments
df_time2['Grouped_Hour']=df_time2['Grouped_Hour'].replace([1,2,3,4,5,6],6)
df_time2['Grouped_Hour']=df_time2['Grouped_Hour'].replace([7,8,9,10,11,12],12)
df_time2['Grouped_Hour']=df_time2['Grouped_Hour'].replace([13,14,15,16,17,18],18)
df_time2['Grouped_Hour']=df_time2['Grouped_Hour'].replace([19,20,21,22,23,24,0],24)
df_time2
```

```python
#Binning Grouped_Month into seasonal segments (3=spring, 6=summer, 9=fall, 12=winter)
df_time2['Grouped_Month']=df_time2['Grouped_Month'].replace([3,4,5],3)
df_time2['Grouped_Month']=df_time2['Grouped_Month'].replace([6,7,8],6)
df_time2['Grouped_Month']=df_time2['Grouped_Month'].replace([9,10,11],9)
df_time2['Grouped_Month']=df_time2['Grouped_Month'].replace([12,1,2],12)
df_time2
```
