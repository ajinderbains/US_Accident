```python
import pandas as pd
import datetime as dt
```

```python
#Bring in cleaned sample from CSV
df_edit_sample=pd.read_csv("./df_sample_trim.csv")
df_edit_sample
```

```python
df_edit_sample = df_edit_sample.rename(columns={"Unnamed: 0": "Original_Index"})
df_edit_sample
```

```python
df_edit_sample['Grouped_Severity'] = df_edit_sample['Severity']
df_edit_sample
```

```python
# Replacing Severity level 1 and 2 to 0
df_edit_sample['Grouped_Severity']=df_edit_sample['Grouped_Severity'].replace([1,2],0)
df_edit_sample
```

```python
# Replacing Severity level 3 and 4 to 1
df_edit_sample['Grouped_Severity']=df_edit_sample['Grouped_Severity'].replace([3,4],1)
df_edit_sample
```

```python
df_edit_sample['Start_Time'] = pd.to_datetime(df_edit_sample['Start_Time'])
df_edit_sample
```

```python
df_edit_sample.dtypes
```

```python
#Extract hour from time
df_edit_sample['Hour'] = df_edit_sample['Start_Time'].dt.hour
df_edit_sample
```

```python
df_edit_sample['Grouped_Hour'] = df_edit_sample['Hour']
df_edit_sample
```

```python
#Binning Grouped_Hour into 6 hour increments
df_edit_sample['Grouped_Hour']=df_edit_sample['Grouped_Hour'].replace([1,2,3,4,5,6],6)
df_edit_sample['Grouped_Hour']=df_edit_sample['Grouped_Hour'].replace([7,8,9,10,11,12],12)
df_edit_sample['Grouped_Hour']=df_edit_sample['Grouped_Hour'].replace([13,14,15,16,17,18],18)
df_edit_sample['Grouped_Hour']=df_edit_sample['Grouped_Hour'].replace([19,20,21,22,23,24,0],24)
df_edit_sample
```

```python
#Extract month from time
df_edit_sample['Month'] = df_edit_sample['Start_Time'].dt.month
df_edit_sample
```

```python
df_edit_sample['Grouped_Month'] = df_edit_sample['Month']
df_edit_sample
```

```python
#Binning Grouped_Month into seasonal segments (3=spring, 6=summer, 9=fall, 12=winter)
df_edit_sample['Grouped_Month']=df_edit_sample['Grouped_Month'].replace([3,4,5],3)
df_edit_sample['Grouped_Month']=df_edit_sample['Grouped_Month'].replace([6,7,8],6)
df_edit_sample['Grouped_Month']=df_edit_sample['Grouped_Month'].replace([9,10,11],9)
df_edit_sample['Grouped_Month']=df_edit_sample['Grouped_Month'].replace([12,1,2],12)
df_edit_sample
```
